```python
df = pd.read_csv("orders.csv", parse_dates=["OrderDate"])
df.head()
# Here we tell Pandas to parse the OrderDate column as a proper datetime.
# In Jupyter, df.head() shows the first 5 rows.
```

```python
df[df["Country"] == "USA"].head()
# This shows only the rows where Country is USA.
```

```python
df[(df["Country"] == "USA") & (df["Shipped"] == "Yes")]
# Step 9: Filter with more than one rule & means AND This gives only orders that are from the USA and are shipped.
```

```python
#How would you filter for rows where Country = Canada?

df[df["Country"] == "Canada"]
```

```python
### Step 10: Sort values
df.sort_values("Price").head()
#Sorts the table by `Price` (smallest first).
```

```python
df.sort_values("Price", ascending=False).head()
#Sorts by `Price` (largest first).
```

```python
# How would you sort the table by Quantity, but from largest to smallest?

df.sort_values("Quantity", ascending=False).head()
```

```python
df.describe()
#Quick summary of numbers  Gives you mean, min, max, etc. for all numeric columns.
```

```python
# Step 12: Group and summarize
#Example: revenue per country
df["Total"] = df["Quantity"] * df["Price"] # first, make a Total column
```

```python
df.groupby("Country")["Total"].sum()
#This says:

# groupby("Country")  group rows by country

# ["Total"].sum()  add up the totals for each country
```

```python
df.groupby("Category")["Price"].mean()
```

```python
df.groupby("Category")["Price"].mean()
```

```python
df.groupby("Category")["Price"].sum()
```

```python
df.groupby("Quantity")["Price"].mean()
```

```python
usa_shipped = df[(df["Country"] == "USA") & (df["Shipped"] == "Yes")]
print(usa_shipped)
```

```python
df.sort_values("Total", ascending=False).head(10)
```

```python
#Recipe 7: Pivot Tables

# Problem: Compare revenue by country and category.
pd.pivot_table(
    df,
    index="Country",
    columns="Category",
    values="Total",
    aggfunc="sum",
    fill_value=0
)
```
